Objective:
Visualize patterns and relationships in the Iris Dataset using different charts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Iris.csv')
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Species', data=df)
plt.title("Species Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['SepalLengthCm'], bins=20)
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',hue='Species',data=df)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
sns.pairplot(df,hue='Species')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.drop('Species',axis=1).corr(),annot=True)
plt.title("Feature Correlation")
plt.show()
```

Conclusion

1. Setosa species is clearly distinguishable.

2. Petal Length and Petal Width are strongly correlated.

3. Versicolor and Virginica show some overlap.

4. Visualizations help understand feature relationships effectively.
